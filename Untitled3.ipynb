```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv("/content/titanic-passengers (1).csv",sep=';')
df.head()
```

```python
print(df.isnull().sum())
```

```python
df.drop('Cabin',inplace=True,axis=1)
df.drop('Fare',inplace=True,axis=1)
df.drop('Ticket',inplace=True,axis=1)
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mod,inplace=True)
```

```python
print(df.isnull().sum())
df.head()
```

```python
one_hot=pd.get_dummies(df['Sex'])
df=df.drop('Sex',axis=1)
df= df.join(one_hot)
df
```

```python
surv_dict = {"Survived" : {"Yes":1, "No":0}}
df.replace(surv_dict, inplace=True)
df
```

```python
embarc_dict={"Embarked":{"S":"Southampton","Q":"Queenstown","C":"Cherbourg"}}
df.replace(embarc_dict,inplace=True)
df
```
